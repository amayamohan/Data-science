```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color= 'red', linestyle= '--', marker= '+')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y, color='blue', marker = '+')
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ['A', 'B', 'C', ]
values = [3,7,5]

plt.bar(categories , values ,color =['red', 'green', 'blue'] )
plt.title("Vertical Bar Chart")
plt.show()
```

```python
categories = ['A', 'B', 'C', ]
values = [3,7,5]

plt.barh(categories , values ,color =['purple', 'black', 'pink'] )
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='black', edgecolor='red')
plt.title("Histogram")
plt.show()
```

```python
sizes = [20, 30, 25,25]
labels = ['A', 'B', 'C','D']
plt.pie(sizes, labels= labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
days =[1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating =  [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.stackplot(days , sleeping, eating, working,playing, labels=['Sleeping', 'Eating', 'Working', 'Playing'], colors=['red', 'black', 'grey', 'purple'])
plt.legend(loc='upper left')
plt.title("Stacked Area Plot")
plt.show()
```

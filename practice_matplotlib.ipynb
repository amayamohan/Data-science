```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color= 'red', linestyle= '--', marker= '+')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y, color='blue', marker = '+')
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ['A', 'B', 'C', ]
values = [3,7,5]

plt.bar(categories , values ,color =['red', 'green', 'blue'] )
plt.title("Bar Chart")
plt.show()
```
